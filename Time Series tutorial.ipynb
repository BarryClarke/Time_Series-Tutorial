dtype: object is used when pandas cant quite figure out what type the data is. The above column is actually a date-time type.

```python
# To be more exact regarding the type in this case, use pd.to_datetime() function
pd.to_datetime(df['date'])
```

Now pandas considers each entry as a date-time, rather than a string of objects as before.

See https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html for more functions relating to time series

```python
# Add the date-time column to the original dataset
df['datetime'] = pd.to_datetime(df['date'])
```

## Lecture on simulating a time series in numpy and pandas

See https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html

```python
# Create a date range from pandas
rng = pd.date_range('1/1/2011', periods=72, freq='H') #https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html
rng
```

See https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.date_range.html for the date range function details, parameters, daylight savings, etc.<br>
The above could be used as an index for a dataframe

```python
# Create a dataframe with a the above time series as the index
import numpy as np
```

```python
# Create a dataframe to simulate the number of hospital admissions in any hour over a 72 hr period
# ts = pd.DataFrame(np.random.poisson(10, len(rng)), index=rng, columns=["admissions"])
ts = pd.DataFrame({'admissions': np.random.poisson(10, len(rng)), 'alcohol related': np.random.poisson(4, len(rng))}, index=rng)
```

```python
ts
```

In the above, you need to be specific with time stamps as to when the start time for each hr begins, when it ends, etc

## Lecture to show slicing and aggregrating time series in pandas

You can now perform all the usual pandas functions for statistical analysis. Examples below

### Slicing

```python
# Calculate the mean of the above dataset
ts.mean()
```

```python
# Access a particulr record
ts.loc['2011-01-01 03:00:00']
```

```python
# Access records between paricular times
ts.loc['2011-01-01 03:00:00':'2011-01-01 23:00:00']
```

```python
# Get the mean for just this time window
ts.loc['2011-01-01 03:00:00':'2011-01-01 23:00:00'].mean()
```

### Aggregation

See https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#aggregation

```python
# Aggregate the dataset by day for the mean
ts.resample('D').mean()
```
